```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
import random
import string

def generate_random_string(length):
     letters = string.ascii_letters + string.digits
     return ''.join(random.choice(letters) for i in range(length))

random_string = generate_random_string(18)
print(random_string)
```

```python
path = "/Users/sasankadowarah/Ultimate_QM_MM/VQA/J=1_h=0.1_g_1=0.2_g_2=0.3/"
os.chdir(path)
# 1 layer ansatz.
cost_function_noisy_1 = np.load("cost_function_values_noisy_1.npy")
cost_function_values_noiseless_1 = np.load("cost_function_values_noiseless_1.npy")
# 2 layers ansatz.
cost_function_noisy_2 = np.load("cost_function_values_noisy_2.npy")
cost_function_values_noiseless_2 = np.load("cost_function_values_noiseless_2.npy")
# 3 layers ansatz.
cost_function_noisy_3 = np.load("cost_function_values_noisy_3.npy")
cost_function_values_noiseless_3 = np.load("cost_function_values_noiseless_3.npy")
# 4 layers ansatz.
cost_function_noisy_4 = np.load("cost_function_values_noisy_4.npy")
cost_function_values_noiseless_4 = np.load("cost_function_values_noiseless_4.npy")
# 5 layers ansatz.
cost_function_noisy_5 = np.load("cost_function_values_noisy_5.npy")
cost_function_values_noiseless_5 = np.load("cost_function_values_noiseless_5.npy")
# 6 layers ansatz.
# cost_function_noisy_6 = np.load("cost_function_values_noisy_6.npy")
# cost_function_values_noiseless_6 = np.load("cost_function_values_noiseless_6.npy")
```

```python
# Plot all cost functions to compare noisy and noiseless simulations
plt.figure(figsize=(14, 7))

# Plot noisy cost functions
plt.plot(cost_function_noisy_1, marker="o", linestyle='-', color="darkorange", linewidth=2, markersize=8, label="Noisy 1 layer")
plt.plot(cost_function_noisy_2, marker="o", linestyle='-', color="blue", linewidth=2, markersize=8, label="Noisy 2 layers")
plt.plot(cost_function_noisy_3, marker="o", linestyle='-', color="green", linewidth=2, markersize=8, label="Noisy 3 layers")
plt.plot(cost_function_noisy_4, marker="o", linestyle='-', color="red", linewidth=2, markersize=8, label="Noisy 4 layers")
plt.plot(cost_function_noisy_5, marker="o", linestyle='-', color="purple", linewidth=2, markersize=8, label="Noisy 5 layers")
# plt.plot(cost_function_noisy_6, marker="o", linestyle='-', color="black", linewidth=2, markersize=8, label="Noisy 6 layers")

# Plot noiseless cost functions
plt.plot(cost_function_values_noiseless_1, marker="x", linestyle='--', color="darkorange", linewidth=2, markersize=8, label="Noiseless 1 layer")
plt.plot(cost_function_values_noiseless_2, marker="x", linestyle='--', color="blue", linewidth=2, markersize=8, label="Noiseless 2 layers")
plt.plot(cost_function_values_noiseless_3, marker="x", linestyle='--', color="green", linewidth=2, markersize=8, label="Noiseless 3 layers")
plt.plot(cost_function_values_noiseless_4, marker="x", linestyle='--', color="red", linewidth=2, markersize=8, label="Noiseless 4 layers")
plt.plot(cost_function_values_noiseless_5, marker="x", linestyle='--', color="purple", linewidth=2, markersize=8, label="Noiseless 5 layers")
# plt.plot(cost_function_values_noiseless_6, marker="x", linestyle='--', color="black", linewidth=2, markersize=8, label="Noiseless 6 layers")

plt.xlabel("Number of Iterations", fontsize=25, labelpad=20)
plt.ylabel("Cost Function Value", fontsize=25, labelpad=20)
plt.title(r"Heisenberg model, $L=3, J = 1, h = 0.1, \gamma_{1} = 0.2, \gamma_{2}= 0.3 $", fontsize=22, pad=15)
plt.legend(fontsize=15, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axhline(y=0, color="k", linestyle="--")
ax = plt.gca()
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.tick_params(axis='both', which='major', length=10, labelsize=20)
plt.tight_layout()
plt.savefig("TFIM_noisy_vs_noiseless_cost_functions_1_6_layers_1.png", dpi=600, bbox_inches='tight')
plt.show()
```
